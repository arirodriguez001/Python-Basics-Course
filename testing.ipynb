## Desafío números

Calcular el promedio de puntos de un equipo.
3 puntos por partido ganado.
1 punto por empate.
0 puntos por derrota.


```python
ganados = int(input('Partidos ganados?: '))
empate = int(input("Partidos empatados?:"))
derrota = int(input("Derrotas?:"))
# print(type(ganados))

puntos = ganados*3 + empate
promedio = puntos / (ganados + empate + derrota)
print(promedio)
```
