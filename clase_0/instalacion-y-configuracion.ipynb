# Tutorial — Instalación de Python y configuración de Google Colab

## Instalación de Python

---

## Python con Windows

### Paso 1 — Instalamos Python con Windows
- Ingresamos a Microsoft Store y buscamos Python.

### Paso 2 — Instalamos Python con Windows
- Lo descargamos.

### Paso 3 — Instalamos Python con Windows
- Una vez termine la instalación:
  - abrimos Windows
  - escribimos `cmd` y enter
  - al escribir `python --version` veremos qué versión se instaló
  - al escribir `python` entraremos a su línea de comando

---

## Python con Mac OS

### Paso 1 — Instalamos Python con MacOS
- Descargaremos Python desde su web oficial: python.org

### Paso 2 — Instalamos Python con MacOS
- En la página principal:
  - ponemos el cursor sobre Downloads
  - en el menú que se despliega pulsamos en Python 3.8.0
  - descargamos “Download for Mac OS X”
  - comenzará automáticamente la descarga

### Paso 3 — Instalamos Python con MacOS
- Una vez terminada la descarga:
  - ejecutamos el archivo `.pkg`
  - se iniciará el asistente de instalación
  - pulsar en continuar
  - se abrirá la pantalla “leeme” con información (versión de Python, sistema, instalación, etc.)
  - pulsar en continuar

### Paso 4 — Instalamos Python con MacOS
- La siguiente pantalla muestra el contrato de licencia
- Al pulsar en Continuar preguntará si lo aceptamos
- Debemos aceptarlo para continuar
- Pulsar en “Acepto”

### Paso 5 — Instalamos Python con MacOS
- Seleccionar el destino donde se instalará Python
- Seleccionamos el disco y pulsamos Continuar

### Paso 6 — Instalamos Python con MacOS
- Elegir tipo de instalación
- Por defecto es instalación estándar (la que se usa en el curso)
- Pulsar en Instalar

### Paso 7 — Instalamos Python con MacOS
- Antes de comenzar:
  - pedirá credenciales del superusuario (privilegios necesarios)
  - introducir usuario y contraseña
  - pulsar “Instalar software”

### Paso 8 — Instalamos Python con MacOS
- Si todo va bien:
  - mensaje indicando que terminó satisfactoriamente
  - quedará instalado Python 3.8.0
  - y el IDE oficial de Python (IDLE) en Aplicaciones/Python 3.8
