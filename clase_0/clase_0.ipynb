# Clase 0 — Introducción a programar con Python

## Objetivo
Entender qué es programar, la diferencia entre frontend y backend, qué es un framework y cómo vamos a trabajar en el curso.

## ¿Qué es programar?

Programar es escribir instrucciones para que una computadora ejecute tareas.

Estas instrucciones:
- Deben seguir reglas estrictas
- Se escriben en un lenguaje de programación

## Evolución de la programación

- Binario (0 y 1)
- Lenguaje ensamblador
- Lenguajes de alto nivel

Python es un lenguaje de alto nivel.
Está más cerca del lenguaje humano.

## Frontend y Backend

### Frontend

- Parte visible de una aplicación
- Interacción con el usuario
- Corre del lado del cliente

Tecnologías:
- HTML
- CSS
- JavaScript

### Backend

- Lógica del sistema
- Corre del lado del servidor
- Maneja datos y reglas de negocio

Lenguajes frecuentes:
- Python
- Java
- Ruby
- PHP

## Funciones del Backend

- Gestión de bases de datos
- Autenticación y permisos
- Comunicación con frontend
- Integración de APIs
- Seguridad
- Testing
- Debugging
- Deploy y mantenimiento

## Frameworks

Un framework es un conjunto de herramientas que:
- Define una estructura
- Facilita el desarrollo
- Evita repetir soluciones comunes

## Django

Django es un framework web para Python.

Permite:
- Crear aplicaciones web completas
- Manejar usuarios
- Administrar bases de datos
- Organizar proyectos de forma profesional

## Rol del programador

Un programador:
- Analiza problemas
- Diseña soluciones
- Escribe código
- Prueba
- Corrige errores
- Mantiene sistemas

## Metodología de estudio recomendada

- Comprender la lógica, no memorizar
- Practicar de forma constante
- Pedir ayuda a tiempo
- Aceptar el error como parte del aprendizaje

## Metodologías ágiles

Las metodologías ágiles permiten:
- Adaptarse a cambios
- Trabajar de forma incremental
- Mejorar continuamente

## Extreme Programming (XP)

XP propone:
- Desarrollo incremental
- Refactorización constante
- Corrección continua de errores
- Simplicidad
- Mejora progresiva


## Dato e información

Dato:
- Valor aislado
- Sin contexto

Información:
- Dato con significado
- Permite tomar decisiones
