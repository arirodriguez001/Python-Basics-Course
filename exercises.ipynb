3 puntos por partido ganado.
1 punto por empate.

Exercise 1: Calculate the multiplication and sum of two numbers: Given two integer numbers, write a Python program to return their product only if the product is equal to or lower than 1000. Otherwise, return their sum.


```python
# given 1:

number1 = 20
number2 = 30

total = number1 * number2
if total <= 1000:
    print(f'el producto es: {total}')
else:
    print(f'la suma es : {number1 + number2}')
```

```python
def multiplicacion_o_suma(num1, num2):
    product = num1 * num2
    if product <= 1000:
        return product
    else:
        return num1 + num2
    
result = multiplicacion_o_suma(20, 30)
print('The result is, ', result)

result2 = multiplicacion_o_suma(40, 30)
print('The result is, ', result)
```

Exercise 4: Remove first n characters from a string:

Write a Python code to remove characters from a string from 0 to n and return a new string.



```python
def remove_chars(word, n):
    print(f'palabra original: {word}')
    if n<= len(word):
        x = word[n:]
        return f'palabra recortada: {x}'
    else:
        return 'palabra recortada: '
        

print('Removiendo caracteres de una cadena')
print(remove_chars('pynative', 4))
print(remove_chars('pynative', 10))
```

Exercise 7: Find the number of occurrences of a substring in a string:

Write a Python code to find how often the substring “Emma” appears in the given string.

```python
str_x = "Emma is good developer. Emma is a writer"

cnt = str_x.count("Emma")
print(cnt)
```

```python
def count_emma(statement):
    print("Given String: ", statement)
    count = 0
    for i in range(len(statement) - 1):
        count += statement[i: i + 4] == 'Emma'
    return count

count = count_emma("Emma is good developer. Emma is a writer")
print("Emma appeared ", count, "times")
```
